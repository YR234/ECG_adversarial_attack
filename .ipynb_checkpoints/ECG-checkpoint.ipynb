```python
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff 
import matplotlib.pyplot as plt
import keras
import sklearn
```

## class 0 is for noramk and 1 for heart attack

```python
def read_data():
    train_path = "./data/ECG200_TRAIN.arff"
    test_path = "./data/ECG200_TEST.arff"
    file = open(train_path, encoding="utf8")
    raw_data = loadarff(file)
    train = pd.DataFrame(raw_data[0])
    train["target"] = train["target"].apply(lambda x: 1 if "-" in str(x) else 0)
    train_labels = train["target"]
    normals = train[train.target == 0]
    abnormal = train[train.target == 1]
    del normals["target"]
    del abnormal["target"]
    del train["target"]
    normals = normals.values
    abnormals = abnormal.values
    
    file = open(test_path, encoding="utf8")
    raw_data = loadarff(file)
    test = pd.DataFrame(raw_data[0])
    test["target"] = test["target"].apply(lambda x: 1 if "-" in str(x) else 0)
    test_labels = test["target"]
    del test["target"]
    
    
    return train.values.reshape(train.shape[0],train.shape[1], 1), keras.utils.to_categorical(train_labels), test.values.reshape(test.shape[0],test.shape[1], 1), keras.utils.to_categorical(test_labels), normals, abnormals
```

## plot normal waves

```python
x_train, y_train, x_test, y_test, normals, abnormals = read_data()
for i, tup in enumerate(zip(normals,abnormals)):
    plt.subplot(321)
    plt.title("Normal")
    plt.plot(list(range(96)), tup[0])

    plt.subplot(322)
    plt.title("AbNormal")
    plt.plot(list(range(96)), tup[1])
    plt.show()
    if i == 10:
        break
```

## Modeling and training

```python
def build_model(input_shape, nb_classes, pre_model=None):
    input_layer = keras.layers.Input((96,1))

    conv1 = keras.layers.Conv1D(filters=128, kernel_size=8, padding='same', input_shape = (96,1))(input_layer)
    conv1 = keras.layers.normalization.BatchNormalization()(conv1)
    conv1 = keras.layers.Activation(activation='relu')(conv1)
    conv1 = keras.layers.Dropout(0.3)(conv1)

    conv2 = keras.layers.Conv1D(filters=256, kernel_size=5, padding='same')(conv1)
    conv2 = keras.layers.normalization.BatchNormalization()(conv2)
    conv2 = keras.layers.Activation('relu')(conv2)
    conv2 = keras.layers.Dropout(0.3)(conv2)


    conv3 = keras.layers.Conv1D(128, kernel_size=3,padding='same')(conv2)
    conv3 = keras.layers.normalization.BatchNormalization()(conv3)
    conv3 = keras.layers.Activation('relu')(conv3)
    conv3 = keras.layers.Dropout(0.3)(conv3)


    gap_layer = keras.layers.pooling.GlobalAveragePooling1D()(conv3)

    output_layer = keras.layers.Dense(nb_classes, activation='softmax')(gap_layer)

    model = keras.models.Model(inputs=input_layer, outputs=output_layer)

    model.compile(loss='categorical_crossentropy', optimizer = keras.optimizers.Adam(),
    metrics=['accuracy'])

    return model

model = build_model(96,2)
model.summary()
```

```python
x_train, y_train, x_test, y_test, normals, abnormals = read_data()
his = model.fit(x_train,y_train, epochs=100)
```

```python
preds = model.predict(x_test)
acc =  sklearn.metrics.accuracy_score(preds.argmax(axis=1), y_test.argmax(axis=1))
print("Test accuracy Score: {}".format(acc))
```
